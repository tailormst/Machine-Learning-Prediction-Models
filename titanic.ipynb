#Data Cleaning

```python
plt.figure(figsize=(12, 7))
sns.boxplot(data=df, x="Pclass", y="Age", palette='winter')
```

```python
def impute_age(cols):
  Age = cols[0]
  Pclass = cols[1]

  if pd.isnull(Age):

    if Pclass == 1:
      return 38
    elif Pclass == 2:
      return 29
    else:
      return 24

  else:
    return Age;

def impute_embark(cols):
  Embarked = cols[0]
  Pclass = cols[1]

  if pd.isnull(Embarked):

    if Pclass == 1:
      return 'S'
    elif Pclass == 2:
      return 'C'
    else:
      return 'C'

  else:
    return Embarked;
```

```python
df['Age'] = df[['Age', 'Pclass']].apply(impute_age, axis=1)
df['Embarked'] = df[['Embarked', 'Pclass']].apply(impute_embark, axis=1)
```

```python
# avoid y-axis
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap="viridis")
## purple = false
## yellow = true
```

```python
df.drop('Cabin', axis=1, inplace=True, errors='ignore')
```

# Converting Categorical Features

```python
df.info()
```

```python
df.head()
```

```python
pd.get_dummies(df['Embarked'], drop_first=True, dtype=int)
```

```python
sex = pd.get_dummies(df['Sex'], drop_first=True, dtype=int)
embark = pd.get_dummies(df['Embarked'], drop_first=True, dtype=int)
```

```python
df.drop(['PassengerId', 'Name', 'Sex','Ticket', 'Embarked'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df = pd.concat([df, sex, embark], axis=1)
df.head()
```

# Split Data Now

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X = df.drop('Survived', axis=1)
y = df['Survived']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
logmodel = LogisticRegression()
logmodel.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred = logmodel.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
```
